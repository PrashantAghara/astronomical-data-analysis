#### Control Flow

```python
for idx in [range(5), range(10)]:
  for i in idx:
    print(i)
  print(idx)
```

#### Functions

```python
def f(x):
  print(x)

f(2)
```

```python
def greet(name, msg):
  print(f'Hey {name}, {msg}')

# Positional Args
greet("Prashant", "How are you ?")
```

```python
# Keyword Args
greet(msg = "How are you ?", name = "Prashant")
```

```python
# Default Args
def greet(name, msg = "Welcome to the team!"):
  print(f'Hey {name}, {msg}')
```

```python
greet("Prashant")
```

```python
def basic_calc(num_1, num_2, operation):
  if operation == 'ADD':
    return num_1 + num_2
  elif operation == 'SUB':
    return num_1 - num_2
  elif operation == 'MULTIPLY':
    return num_1 * num_2
  elif operation == 'DIVIDE':
    if num_2 == 0:
      return "ERROR"
    else:
      return num_1 / num_2
  else:
    return "Invalid Operation"
```

```python
print(basic_calc(1, 1, "ADD"))
print(basic_calc(1, 1, "DIVIDE"))
print(basic_calc(1, 2, "SUB"))
print(basic_calc(1, 3, 'DIVIDE'))
print(basic_calc(1, 1, "POW"))
```
